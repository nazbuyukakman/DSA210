```python
import pandas as pd
import numpy as np

path = "DSA210-dataset.csv"


raw = pd.read_csv(path, header=None, sep=None, engine="python")


first_cell = str(raw.iloc[0, 0]).strip().upper()

if first_cell in ["DATES", "DATE"]:
   
    df = raw.iloc[1:, [0, 1, 2]].copy()
else:
   
    df = raw.iloc[:, [0, 1, 2]].copy()

df.columns = ["date", "sweets_consumed", "on_period"]


df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")
df["sweets_consumed"] = pd.to_numeric(df["sweets_consumed"], errors="coerce")
df["on_period"] = pd.to_numeric(df["on_period"], errors="coerce")

df = df.dropna(subset=["date", "sweets_consumed", "on_period"])
df["sweets_consumed"] = df["sweets_consumed"].astype(int)
df["on_period"] = df["on_period"].astype(int)

df = df.sort_values("date").reset_index(drop=True)

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

X = df[["on_period"]]
y = df["sweets_consumed"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

y_pred_lr = log_reg.predict(X_test)
print("LOGISTIC REGRESSION")
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
tree = DecisionTreeClassifier(max_depth=2, random_state=42)
tree.fit(X_train, y_train)

y_pred_tree = tree.predict(X_test)
print("DECISION TREE")
print(confusion_matrix(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
```

```python
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
print("RANDOM FOREST")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
results = {
    "Logistic Regression": log_reg.score(X_test, y_test),
    "Decision Tree": tree.score(X_test, y_test),
    "Random Forest": rf.score(X_test, y_test)
}

results
```
